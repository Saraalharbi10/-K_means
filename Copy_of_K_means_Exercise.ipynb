#Step 3 : Elbow Method
Finding the ideal number of groups to divide the data into is a basic stage in any unsupervised algorithm. One of the most common techniques for figuring out this ideal value of k is the elbow approach.





```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


iris = load_iris()
X = iris.data


k_range = range(1, 11)
inertia = []


for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

#Step 4 : Plot the Elbow graph to find the optimum number of cluster



The plot displays a scatter plot of data points (X[:,0], X[:,1]) with grid lines. It also marks the initial cluster centers (red stars) generated for K-means clustering.



```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


iris = load_iris()
X = iris.data


k_range = range(1, 11)
inertia = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o', linestyle='--', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.grid(True)

k = 3
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(X)
centers = kmeans.cluster_centers_
labels = kmeans.labels_

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=50, alpha=0.7, edgecolors='k')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='*', edgecolors='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title(f'K-Means Clustering (k={k})')
plt.grid(True)

plt.tight_layout()
plt.show()
```

#Step 5 : Build the Kmeans clustering model



*   Find the cluster center
*   Predict the cluster group





```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


iris = load_iris()
X = iris.data


k = 3


kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(X)


centers = kmeans.cluster_centers_
print("Cluster Centers:\n", centers)


labels = kmeans.predict(X)
print("Predicted Labels:\n", labels)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=50, alpha=0.7, edgecolors='k')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='*', edgecolors='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title(f'K-Means Clustering with k={k}')
plt.grid(True)
plt.show()
```

#Step 6 : Predict the cluster group

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


iris = load_iris()
X = iris.data

k = 3
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(X)

labels = kmeans.predict(X)


print("Predicted Labels:\n", labels)


plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=50, alpha=0.7, edgecolors='k')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title(f'K-Means Clustering with k={k}')
plt.grid(True)
plt.show()
```
